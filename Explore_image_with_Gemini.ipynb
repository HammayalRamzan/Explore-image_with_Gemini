```python
# API Key setup
from google.colab import userdata
GEMINI_API_KEY = userdata.get('GOOGLE_API_KEY')
```

```python
# Configure the genai library and load the model you want to use.
from google import genai
from google.genai import Client

client: Client = genai.Client(
    api_key=GEMINI_API_KEY,
)

model: str = "gemini-2.0-flash-exp"
```

```python
# Generate content using the model
from google.genai.types import GenerateContentResponse
from IPython.display import display, Markdown

response: GenerateContentResponse = client.models.generate_content(
    model=model,
    contents='How does AI work?'
)
```

```python
display(response)
```

```python
# Use markdown to generate response
display(Markdown(response.text))
```

```python
# Generate content using tokens
from google.genai.types import CountTokensResponse
result: CountTokensResponse = client.models.count_tokens(
    model=model,
    contents="What do you know about AI?",
)

print(result)
```

```python
# Import URL of image
import PIL.Image
img = PIL.Image.open('/content/Faisal-Fayyaz.jpg')
display(img)
```

```python
# Summarize the picture in text
prompt='summarize the picture'
```

```python
# Summarize the picture in text using markdown
response: GenerateContentResponse = client.models.generate_content(
    model=model,
    contents=[img,prompt]
)
display(Markdown(response.text))
```

```python
# Recall the LLM my name
response: GenerateContentResponse = client.models.generate_content(
    model=model,
    contents='Hey, My name is Hammayal Ramzan'
)
display(Markdown(response.text))
```

```python
# Ask the LLM what my name using Markdown
response: GenerateContentResponse = client.models.generate_content(
    model=model,
    contents='Hey, What is my name'
)
display(Markdown(response.text))
```

```python
from google.genai.chats import Chat

# Initiate a chat
chat: Chat = client.chats.create(
    model=model
)

# Send chat a message
response: GenerateContentResponse = chat.send_message("Hey, My name is Hammayal Ramzan")
Markdown(response.text)
```

```python
# Again ask to LLM what my name
response: GenerateContentResponse = chat.send_message("What is my name?")
Markdown(response.text)
```
